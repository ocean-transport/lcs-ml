```python
import numpy as np
import xarray as xr
import dask
from os import listdir
```

```python
dir = listdir('/burg/abernathey/users/hillary/') 
```

```python
EM = xr.open_mfdataset('/burg/abernathey/users/hillary/QG_proto_EM_*.nc', concat_dim='ensemble_member', combine='by_coords')
EM.dims
```

```python
def corr_PV(EM):
    corr = xr.corr(EM.q.isel(lev=0, ensemble_member=1),EM.q.isel(lev=0, ensemble_member=2), dim=('x','y'))
    return corr

corr = corr_PV(EM)
corr.plot();
```
